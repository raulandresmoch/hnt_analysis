```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
 
```

```python
abril_df = pd.read_csv(r"C:\Users\Raul\Documents\Reportes Helium\heliumtracker-report-2022-4.csv")
mayo_df = pd.read_csv(r"C:\Users\Raul\Documents\Reportes Helium\heliumtracker-report-2022-5.csv")
junio_df = pd.read_csv(r"C:\Users\Raul\Documents\Reportes Helium\heliumtracker-report-2022-6.csv")
julio_df = pd.read_csv(r"C:\Users\Raul\Documents\Reportes Helium\heliumtracker-report-2022-7.csv")
```

```python
union = [abril_df,mayo_df,junio_df,julio_df]
reporte_general = pd.concat(union) 
reporte_general=reporte_general.drop(columns=['Glorious Mossy Hare','HNT value in USD','HNT Value in USD'])
reporte_general['Date']=pd.to_datetime(reporte_general['Date'])
reporte_general=reporte_general.reset_index()
reporte_general=reporte_general.drop(columns=['index'])
```

```python
promedio_dia=reporte_general.mean()
promedio_dia=promedio_dia.reset_index()
promedio_dia=promedio_dia.rename(columns={'index':'minero',0:'Avg_Dia'})
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='minero',y='Avg_Dia',data=promedio_dia)
plt.xticks(rotation='vertical')
plt.show()
```
